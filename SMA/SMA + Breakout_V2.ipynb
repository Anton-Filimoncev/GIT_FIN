```python
weights = np.arange(1,11)

def plot_returns_dd(portfolio):
    portfolio = portfolio.copy()
    # ----------- Sharpe ratio ------------------
    sharpe_ratio = np.mean(portfolio.str_returns)/np.std(portfolio.str_returns)*(252**0.5)
    # ----------- Cumulative strategy returns ------------------
    portfolio['cum_str_returns'] = (portfolio['str_returns']+1).cumprod()
    # ----------- Drawdown ------------------    
    # Calculate the running maximum
    running_max = np.maximum.accumulate(portfolio['cum_str_returns'].dropna())
    running_max[running_max < 1] = 1
    drawdown = (portfolio['cum_str_returns'])/running_max - 1
    max_dd = drawdown.min()*100
    cumulative_return= (portfolio['cum_str_returns'].iloc[-2] - 1) * 100
      
    return sharpe_ratio, max_dd, cumulative_return



def plot_returns_dd_next(portfolio, returns):
    portfolio = portfolio.copy()
    # ----------- Sharpe ratio ------------------
    sharpe_ratio = np.mean(portfolio[returns])/np.std(portfolio[returns])*(252**0.5)
    # ----------- Cumulative strategy returns ------------------
    portfolio['cum_str_returns'] = (portfolio[returns]+1).cumprod()
    # ----------- Drawdown ------------------    
    # Calculate the running maximum
    running_max = np.maximum.accumulate(portfolio['cum_str_returns'].dropna())
    running_max[running_max < 1] = 1
    drawdown = (portfolio['cum_str_returns'])/running_max - 1
    max_dd = drawdown.min()*100
    cumulative_return= (portfolio['cum_str_returns'].iloc[-2] - 1) * 100
      
    return sharpe_ratio, max_dd, cumulative_return



def plot_returns_dd_index(portfolio_index):
    portfolio_index = portfolio_index.copy()
    portfolio_index['str_returns'] = portfolio_index['returns'].shift(-1)
    sharpe_ratio = np.mean(portfolio_index.str_returns)/np.std(portfolio_index.str_returns)*(252**0.5)
    portfolio_index['cum_str_returns'] = (portfolio_index['str_returns']+1).cumprod()
        # ----------- Drawdown ------------------    
    # Calculate the running maximum
    running_max = np.maximum.accumulate(portfolio_index['cum_str_returns'].dropna())
    running_max[running_max < 1] = 1
    drawdown = (portfolio_index['cum_str_returns'])/running_max - 1
    max_dd = drawdown.min()*100     
    cumulative_return = (portfolio_index['cum_str_returns'].iloc[-2] - 1) * 100
    
    return sharpe_ratio, max_dd, cumulative_return
    
def get_strategy_returns_breakout(portfolio_breakout):
    portfolio_breakout = portfolio.copy()
#     print(portfolio_breakout)
    # Calculate the breakout indicator values
    portfolio_breakout['high'] = portfolio_breakout.value.rolling(window=3).max()
    # Create a trading signal
    portfolio_breakout['signal'] = np.where(portfolio_breakout.value>=portfolio_breakout.high,1,0)
    # Calculate the strategy returns
    portfolio_breakout['str_returns'] = portfolio_breakout['returns'].shift(-1) * portfolio_breakout['signal']
#     print(portfolio_breakout)
    return portfolio_breakout 
    
def get_strategy_returns_sma_breakout(portfolio_sma_breakout):
    portfolio_sma_breakout = portfolio_sma_breakout.copy()
#     print(portfolio_sma_breakout)
    # Calculate the simple moving average 
    sma10 = portfolio_sma_breakout.value >  portfolio.value.rolling(window=10).mean()
    # Calculate the breakout indicator values
    breakout = portfolio_sma_breakout.value>=portfolio_sma_breakout.value.rolling(window=3).max()
    # Create a trading signal
    portfolio_sma_breakout['signal'] = np.where(sma10 & breakout,1,0)
    # Calculate the strategy returns
    portfolio_sma_breakout['str_returns'] = portfolio_sma_breakout['returns'].shift(-1) * portfolio_sma_breakout['signal']
#     print(portfolio_sma_breakout)
    
    return portfolio_sma_breakout
    
    
def get_strategy_returns_sma(portfolio_sma):
    portfolio_sma = portfolio_sma.copy()
#     print(portfolio_sma)
    # Calculate the simple moving average of period 10
    portfolio_sma['sma10'] =  portfolio.value.rolling(window=10).mean()
    # Create a trading signal
    portfolio_sma['signal'] = np.where(portfolio_sma.value > portfolio_sma.sma10,1,0)
    print('^'*50)
    print(portfolio_sma)
    print('^'*50)
    # Calculate the strategy returns
    portfolio_sma['str_returns'] = portfolio_sma['returns'].shift(-1) * portfolio_sma['signal']
#     print(portfolio_sma)
#     print(print('+'*50))
    return portfolio_sma


def get_strategy_returns_ema_breakout(portfolio_ema_breakout):
    portfolio_ema_breakout = portfolio_ema_breakout.copy()
#     print(portfolio_sma_breakout)
    # Calculate the simple moving average 
    sma10 = portfolio_ema_breakout.value >  pd.Series.ewm(portfolio['value'], span=10).mean()
    # Calculate the breakout indicator values
    breakout = portfolio_ema_breakout.value>=portfolio_ema_breakout.value.rolling(window=3).max()
    # Create a trading signal
    portfolio_ema_breakout['signal_ema'] = np.where(sma10 & breakout,1,0)
    # Calculate the strategy returns
    portfolio_ema_breakout['str_returns_ema'] = portfolio_ema_breakout['returns'].shift(-1) * portfolio_ema_breakout['signal_ema']
#     print(portfolio_sma_breakout)
    
    return portfolio_ema_breakout
    
    
def get_strategy_returns_ema(portfolio_ema):
    portfolio_ema = portfolio_ema.copy()
#     print(portfolio_sma)
    # Calculate the simple moving average of period 10
    portfolio_ema['ema10'] =  pd.Series.ewm(portfolio['value'], span=10).mean()
    # Create a trading signal
    portfolio_ema['signal_ema'] = np.where(portfolio_ema.value > portfolio_ema.ema10,1,0)
    # Calculate the strategy returns
    portfolio_ema['str_returns_ema'] = portfolio_ema['returns'].shift(-1) * portfolio_ema['signal_ema']
#     print(portfolio_sma)
#     print(print('+'*50))
    return portfolio_ema

def get_strategy_returns_wma_breakout(portfolio_wma_breakout):
    portfolio_wma_breakout = portfolio_wma_breakout.copy()
#     print(portfolio_sma_breakout)
    # Calculate the simple moving average 
    sma10 = portfolio_wma_breakout.value >  portfolio_wma_breakout.value.rolling(10).apply(lambda prices: np.dot(prices, weights)/weights.sum(), raw=True)
    # Calculate the breakout indicator values
    breakout = portfolio_wma_breakout.value>=portfolio_wma_breakout.value.rolling(window=3).max()
    # Create a trading signal
    portfolio_wma_breakout['signal_wma'] = np.where(sma10 & breakout,1,0)
    # Calculate the strategy returns
    portfolio_wma_breakout['str_returns_wma'] = portfolio_wma_breakout['returns'].shift(-1) * portfolio_wma_breakout['signal_wma']
#     print(portfolio_sma_breakout)
    
    return portfolio_wma_breakout
    
def get_strategy_returns_wma(portfolio_wma):
    portfolio_wma = portfolio_wma.copy()
#     print(portfolio_sma)
    # Calculate the simple moving average of period 10
    portfolio_wma['wma10'] =  portfolio_wma.value.rolling(10).apply(lambda prices: np.dot(prices, weights)/weights.sum(), raw=True)
    # Create a trading signal
    portfolio_wma['signal_wma'] = np.where(portfolio_wma.value > portfolio_wma.wma10,1,0)
    # Calculate the strategy returns
    portfolio_wma['str_returns_wma'] = portfolio_wma['returns'].shift(-1) * portfolio_wma['signal_wma']
#     print(portfolio_sma)
#     print(print('+'*50))
    return portfolio_wma

itteration_result = pd.DataFrame()
main_result = pd.DataFrame()

cumSma = []
cumBrake = []
cumSmaBrake = []

cumEma = []
cumEmaBrake = []

max_ddEma = []
max_ddEmaBrake = []
sharpeEma = []
sharpeEmaBrake = []

max_ddWma = []
max_ddWmaBrake = []
sharpeWma = []
sharpeWmaBrake = []

cumWma = []
cumWmaBrake = []

max_ddSma = []
max_ddBrake = []
max_ddSmaBrake = []

sharpeSma = []
sharpeBrake = []
sharpeSmaBrake = []

cumIndex = []
max_ddIndex = []
sharpeIndex = []

tik_buy = 0

start_year = 2010

# Calculate the standard deviation
# data_std = np.std(p, axis = 0)
for i in range(2021 - start_year): #
    data_std = data_pc.loc[f'{start_year+i}-01-01':f'{start_year+i}-12-31'].std()*(252**0.5)*100
#     data_std = data_pc.loc['2020-05-27':'2021-05-26'].std()*(252**0.5)*100
#     print(data_std)

#     print('Date:', start_year+i, 'STD:', data_std)
    vol_sorted = data_std.sort_values(ascending=False)
    top_decile = vol_sorted[:int(len(data_std)*decile_coef)]
#     print('top_decile', top_decile)
    data = p
    data_index =  index_ticker_p
    
    data['Date'] = data.index.tolist()
    data_index['Date'] = data_index.index.tolist()
    
    data = data.set_index(data['Date'].dt.year)
    data_index = data_index.set_index(data_index['Date'].dt.year)
    
    # Retrieve data in stock_list from 2019 January onwards
    stock_data = data.loc[start_year+i+1,top_decile.index]
#     stock_data = data.loc['2021-4':'2021-6',top_decile.index] 
#     print(stock_data)
    
    if i == (2021-start_year-1):
#         print(top_decile.index)
        tik_buy= top_decile.index.tolist()
    stock_data_index = data_index.loc[start_year+i+1,index_ticker]

    # Calculate the daily percentage change of prices
    
    stock_data_pc = stock_data.pct_change()
    
#     stock_data_pc.replace([np.inf, -np.inf], np.nan, inplace=True)
#     print(stock_data)
    
#     print(stock_data_pc)
    
    stock_data_index = stock_data_index.pct_change()
    # Create a new dataframe called portfolio
    portfolio = pd.DataFrame()
    portfolio_index = pd.DataFrame()
    
    # Calculate the average returns of stocks
    portfolio['returns'] = stock_data_pc.mean(axis=1)
#     print(portfolio)
    portfolio_index['returns'] = stock_data_index.mean(axis=1)
    # Calculate cumulative returns of portfolio
    portfolio['value'] = (portfolio+1).cumprod()
    portfolio_index['value'] = (portfolio_index+1).cumprod()
    # Drop any rows with nan values
    portfolio = portfolio.dropna()
    
#     print(portfolio)
    portfolio_index = portfolio_index.dropna()

    
    
    
    portfolio_sma = get_strategy_returns_sma(portfolio)  
    portfolio_ema = get_strategy_returns_ema(portfolio)  
    portfolio_wma = get_strategy_returns_wma(portfolio)  
    
    portfolio_breakout =  get_strategy_returns_breakout(portfolio)   
    portfolio_sma_breakout = get_strategy_returns_sma_breakout(portfolio) 
    
    portfolio_ema_breakout = get_strategy_returns_ema_breakout(portfolio) 
    portfolio_wma_breakout = get_strategy_returns_wma_breakout(portfolio) 
    
   
    sharpe_ratio_sma, max_dd_sma, cumulative_return_sma = plot_returns_dd(portfolio_sma)
    sharpe_ratio_ema, max_dd_ema, cumulative_return_ema = plot_returns_dd_next(portfolio_ema, 'str_returns_ema')
    sharpe_ratio_wma, max_dd_wma, cumulative_return_wma = plot_returns_dd_next(portfolio_wma, 'str_returns_wma')
    
    sharpe_ratio_breakout, max_dd_breakout, cumulative_return_breakout = plot_returns_dd(portfolio_breakout)
    
    sharpe_ratio_sma_breakout, max_dd_sma_breakout, cumulative_return_sma_breakout = plot_returns_dd(portfolio_sma_breakout)
    sharpe_ratio_ema_breakout, max_dd_ema_breakout, cumulative_return_ema_breakout = plot_returns_dd_next(portfolio_ema_breakout, 'str_returns_ema')
    sharpe_ratio_wma_breakout, max_dd_wma_breakout, cumulative_return_wma_breakout = plot_returns_dd_next(portfolio_wma_breakout, 'str_returns_wma')
        
        
    sharpe_ratio_index, max_dd_index, cumulative_return_index = plot_returns_dd_index(portfolio_index)
    
    cumEma.append(cumulative_return_ema)
    cumEmaBrake.append(cumulative_return_ema_breakout)
    
    max_ddEma.append(max_dd_ema)
    max_ddEmaBrake.append(max_dd_ema_breakout)
    
    sharpeEma.append(sharpe_ratio_ema)
    sharpeEmaBrake.append(sharpe_ratio_ema_breakout)
    
    max_ddWma.append(max_dd_wma)
    max_ddWmaBrake.append(max_dd_wma_breakout)
    
    sharpeWma.append(sharpe_ratio_wma)
    sharpeWmaBrake.append(sharpe_ratio_wma_breakout)
    
    cumWma.append(cumulative_return_wma)
    cumWmaBrake.append(cumulative_return_wma_breakout)
        
    cumSma.append(cumulative_return_sma)
    cumBrake.append(cumulative_return_breakout)
    cumSmaBrake.append(cumulative_return_sma_breakout)

    max_ddSma.append(max_dd_sma)
    max_ddBrake.append(max_dd_breakout)
    max_ddSmaBrake.append(max_dd_sma_breakout)

    sharpeSma.append(sharpe_ratio_sma)
    sharpeBrake.append(sharpe_ratio_breakout)
    sharpeSmaBrake.append(sharpe_ratio_sma_breakout)
    
    cumIndex.append(cumulative_return_index)
    max_ddIndex.append(max_dd_index)
    sharpeIndex.append(sharpe_ratio_index)

    print('*'*50)

main_result['Cum SMA'] = cumSma
main_result['Cum EMA'] = cumEma
main_result['Cum WMA'] = cumWma

main_result['Cum SMA & Brake'] = cumSmaBrake    
main_result['Cum EMA & Brake'] = cumEmaBrake
main_result['Cum WMA & Brake'] = cumWmaBrake    

main_result['Max Drawdown SMA'] = max_ddSma
main_result['Max Drawdown EMA'] = max_ddEma
main_result['Max Drawdown WMA'] = max_ddWma

main_result['Max Drawdown SMA & Brake'] = max_ddSmaBrake
main_result['Max Drawdown EMA & Brake'] = max_ddEmaBrake
main_result['Max Drawdown WMA & Brake'] = max_ddWmaBrake

main_result['Sharpe Ratio SMA'] = sharpeSma
main_result['Sharpe Ratio EMA'] = sharpeEma
main_result['Sharpe Ratio WMA'] = sharpeWma

main_result['Sharpe Ratio SMA & Brake'] = sharpeSmaBrake
main_result['Sharpe Ratio EMA & Brake'] = sharpeEmaBrake
main_result['Sharpe Ratio WMA & Brake'] = sharpeWmaBrake



main_result['Cum Brake'] = cumBrake
main_result['Max Drawdown Brake'] = max_ddBrake
main_result['Sharpe Ratio Brake'] = sharpeBrake

main_result['Index Cum'] = cumIndex
main_result['Index Max Drawdown'] = max_ddIndex
main_result['Index Sharpe Ratio'] = sharpeIndex

table_date = []

start_year = 2010
end = datetime.datetime.now()

start_year = 2010

for k in range(start_year+1, end.year+1):
    table_date.append(k)

# print(table_date)   
main_result['Date'] = table_date

main_result = main_result.set_index(main_result.Date) 
main_result = main_result.T

# main_result = main_result.drop( index='Date')
```
